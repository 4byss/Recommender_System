```python
data = {
  "Syn": []
}

df1 = pd.DataFrame(data)

df1['Syn'] = df['Synopsis']

df1.head(5)

df['Syn'] = df1
df.head(5)
# df.join(df1)
# pd.concat([df,df1])
# df.head(5)
```

```python
df['Synopsis'] = df['Synopsis'].apply(lambda x:x.split(' '))
df['English'] = df['English'].apply(lambda x:x.split(','))
df['Status'] = df['Status'].apply(lambda x:x.split(','))
df['Studios'] = df['Studios'].apply(lambda x:x.split(','))
df['Source'] = df['Source'].apply(lambda x:x.split(','))
df['Genres'] = df['Genres'].apply(lambda x:x.split(','))
df['Themes'] = df['Themes'].apply(lambda x:x.split(','))
df['Demographics'] = df['Demographics'].apply(lambda x:x.split())
```

```python
df['English'] = df['English'].apply(lambda x:[i.replace(' ', '') for i in x])
df['Status'] = df['Status'].apply(lambda x:[i.replace(' ', '') for i in x])
df['Studios'] = df['Studios'].apply(lambda x:[i.replace(' ', '') for i in x])
df['Source'] = df['Source'].apply(lambda x:[i.replace(' ', '') for i in x])
df['Genres'] = df['Genres'].apply(lambda x:[i.replace(' ', '') for i in x])
df['Themes'] = df['Themes'].apply(lambda x:[i.replace(' ', '') for i in x])
df['Demographics'] = df['Demographics'].apply(lambda x:[i.replace(' ', '') for i in x])
```

```python
df.head(5)
```

```python
df['Tag'] = df['English'] + df['Synopsis'] + df['Status'] + df['Studios'] + df['Source'] + df['Genres'] + df['Themes'] + df['Demographics']
```

```python
new_df = df[['ID', 'Title', 'Syn', 'Tag']]
```

```python
new_df
```

```python
new_df['Tag'] = new_df['Tag'].apply(lambda x:" ".join(x))
```

```python
new_df.head(5)
```

```python
new_df['Tag'][0]
```

```python
new_df['Tag'][1]
```

```python
new_df['Tag'] = new_df['Tag'].apply(lambda x:x.lower())
```

```python
new_df.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['Tag']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(anime):
    index = new_df[new_df['Title'] == anime].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].Title)
```

```python
recommend('86')
```
